## Data Assesing

```python
def assess_dataframe(df):

    print("Data Assessment Report")

    print(f"\nShape of dataset: {df.shape[0]} Rows x {df.shape[1]} Columns")

    print("\nColumn Information:")
    info_table = pd.DataFrame({
        "DataType": df.dtypes,
        "Non-Null Count": df.notnull().sum(),
        "Null Count": df.isnull().sum(),
        "Unique Values": df.nunique()
    })
    print(info_table)
    
    object_cols = df.select_dtypes(include=["object"]).columns.tolist()
    numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
    print(f"\nObject Columns ({len(object_cols)}): {object_cols if object_cols else 'None'}")
    print(f"Numeric Columns ({len(numeric_cols)}): {numeric_cols if numeric_cols else 'None'}")

    dup_count = df.duplicated().sum()
    print(f"\nDuplicate Rows: {dup_count}")

    print("\nSummary Statistics:")
    print(df.describe().T)

    print("\nData Assessment Completed.")
```

## Text Preprocessing

```python
def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    text = re.sub(r'#[A-Za-z0-9_]+', '', text)
    text = re.sub(r'RT[\s]+', '', text)
    text = re.sub(r"http\S+", '', text)
    text = re.sub(r'[0-9]+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = text.replace('\n', ' ')
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.strip()
    return text

def casefoldingText(text):
    return text.lower()

def tokenizingText(text):
    return word_tokenize(text)

def filteringText(text_tokens):
    listStopwords = set(stopwords.words('indonesian'))
    listStopwords.update(stopwords.words('english'))
    update = {'iya','yaa','gak','nya','na','sih','ku',"di","ga","ya","gaa","loh","kah","woi","woii","woy","gk","njir","eh","yang","dan","hm","hmm"}
    listStopwords.update(update)

    filtered = [word for word in text_tokens if word not in listStopwords]
    return filtered

def stemmingText(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()

    words = text.split()
    stemmed_words = [stemmer.stem(word) for word in words]
    stemmed_text = ' '.join(stemmed_words)

    return stemmed_text

def toSentence(list_words):
    return ' '.join(list_words)
```
